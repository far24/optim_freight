```python
# read data from the input file
with open("input_d2.txt", "r") as f:
    # read line by line
    lines = f.readlines()
```

```python
invalid = 0
for line in lines:
    #print(line)
    
    # split each line by space
    colon_split = line.split(":")
    space_split = colon_split[0].split(" ")
    min_cond = space_split[0].split("-")[0]
    max_cond = space_split[0].split("-")[1]
    letter = space_split[1]
    password = colon_split[1]

    # print(f"find letter {letter} at least {min_cond} times and at most {max_cond} times in {password}")
    # print(f"Number of times Letter: '{letter}' is in {password} is {password.count(letter)}")
    
    check_val = int(password.count(letter))
    
    if check_val >= int(min_cond) and check_val <= int(max_cond):
        print(letter, password, min_cond, max_cond, check_val)
    else:
        print(line, letter, password, min_cond, max_cond, check_val)
        invalid += 1
        print("...............................INVALID......................................")
        
    print(f"Number of invalid passwords: {invalid}")
```
